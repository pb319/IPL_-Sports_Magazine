```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
Bowling = pd.read_excel("Bowling(new).xlsx")
Bowling.head(3)
```

```python
#Scatter Plot Showing Relationship between PC1 and Wkts
Bowling.plot(kind='scatter', x='PC1', y='Wkts', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
#Scatter Plot Showing Relationship between PC2 and Wkts
Bowling.plot(kind='scatter', x='PC2', y='Wkts', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# Split data into features (X) and target variable (y)
X = Bowling[['PC1', 'PC2']]
y = Bowling['Wkts']

# Initialize the linear regression model
modelV = LinearRegression()

# Fit the model using the training data
modelV.fit(X, y)

# Print the coefficients
print("Coefficients:", modelV.coef_)
print("Intercept:", modelV.intercept_)
```
