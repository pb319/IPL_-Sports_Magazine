```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
Bating_100 = pd.read_excel("Bating_100(new).xlsx")
Bating_100.head(3)
```

```python
#Scatter Plot Showing Relationship between PC1 and Runs
Bating_100.plot(kind='scatter', x='PC1', y='Runs', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
#Scatter Plot Showing Relationship between PC2 and Runs
Bating_100.plot(kind='scatter', x='PC2', y='Runs', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
#Scatter Plot Showing Relationship between PC3 and Runs
Bating_100.plot(kind='scatter', x='PC3', y='Runs', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

***- Lets Fit the Model***

```python
# Split data into features (X) and target variable (y)
X = Bating_100[['PC1', 'PC2']]
y = Bating_100['Runs']

# Initialize the linear regression model
modelV = LinearRegression()

# Fit the model using the training data
modelV.fit(X, y)

# Print the coefficients
print("Coefficients:", modelV.coef_)
print("Intercept:", modelV.intercept_)
```
